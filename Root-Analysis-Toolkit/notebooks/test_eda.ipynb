```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import cv2
import numpy as np
```

```python
masks = sorted([
    p for p in Path("data/organised/Y23_npec/mask").rglob("*root_mask.tif")
    if "occluded" not in p.name
])
len(masks)
```

```python
# Check unique pixel values of the mask[0]
mask = cv2.imread(str(masks[0]), cv2.IMREAD_UNCHANGED)

values = mask.flatten()

unique = np.unique(values)
unique
```

```python
for mask in masks[:10]:
    mask_read = cv2.imread(str(mask), cv2.IMREAD_GRAYSCALE)

    flat_mask = mask_read.flatten()

    size = round(sum(flat_mask == 1) / len(flat_mask),6)

    # Plot the mask
    plt.imshow(mask_read, cmap="gray")
    plt.title(mask.name)
    plt.axis("off")
    plt.show()
    print(f"Mask: {mask.name}, Size: {size}")
```
